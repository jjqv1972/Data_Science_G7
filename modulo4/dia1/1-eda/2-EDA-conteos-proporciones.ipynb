# EDA - CONTEOS Y PROPORCIONES

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
sns.set_style(style='whitegrid')
sns.set_context(context='notebook')
plt.rcParams['figure.figsize'] = (11, 9.4)

penguin_color = {
    'Adelie': '#ff6602ff',
    'Gentoo': '#0f7175ff',
    'Chinstrap': '#c65dc9ff'
}



penguins_df = sns.load_dataset('penguins')
penguins_df.dropna(inplace=True)
penguins_df
```

# CONTEOS
## CONTEO DE TODAS LAS VARIABLES

```python
penguins_df.describe(include='all')
```

# CONTEO DE VARIABLES CATEGORICAS

```python
penguins_df.describe(include='object')
```

# CONTEO DE VARIABLES NUMERICAS

```python
penguins_df.describe(include=[np.number])
```

# GRAFICO DE CONTEOS

```python
(
    penguins_df
    .species
    .value_counts()
    .plot(kind='bar',
          color=penguin_color.values(),
          title='Cantidad de Pinguinos por Especie')
)
```

# grafico de conteo por Islas

```python
island_color = {
    'Biscoe': '#ff3302ff',
    'Dream': '#0f7175ff',
    'Torgersen': '#c65dc9ff'
}
```

```python
(
    penguins_df
    .island
    .value_counts()
    .plot(
        kind='bar',
        color=island_color.values(),
        title='Cantidad de Pinguinos por Isla'
    )
)
```

# PROPORCIONES

```python
penguins_df['variable'] = ''
penguins_df
```

```python
(
    penguins_df
    .pipe(
        lambda df:(
            sns.displot(
                data=df,
                x='variable',
                hue='species',
                multiple='fill',
                palette=penguin_color
            )
        )
    )
)
```

# PROPORCIONES POR ISLA

```python
(
    penguins_df
    .pipe(
        lambda df:(
            sns.displot(
                data=df,
                x='variable',
                hue='island',
                multiple='fill',
                palette=island_color
            )
        )
    )
)
```
