```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
sns.set_style(style='whitegrid')
sns.set_context(context='notebook')
plt.rcParams['figure.figsize'] = (11, 9.4)

penguin_color = {
    'Adelie': '#ff6602ff',
    'Gentoo': '#0f7175ff',
    'Chinstrap': '#c65dc9ff'
}

penguins_df = sns.load_dataset('penguins')
penguins_df.dropna(inplace=True)
penguins_df
```

# MEDIA

## MEDIA USANDO NUMPY

```python
x = 'body_mass_g'
```

```python
np.mean(penguins_df[x])
```

## MEDIA USANDO PANDAS

```python
penguins_df[x].mean()
```

## MEDIA DE TODAS LAS VARIABLES USANDO PANDAS

```python
penguins_df.mean(numeric_only=True)
```

# MEDIANA

## MEDIANA USANDO NUMPY

```python
np.median(penguins_df[x])
```

## MEDIANA USANDO PANDAS

```python
penguins_df[x].median()
```

## MEDIANA DE TODAS LAS VARIABLES NUMERICAS

```python
penguins_df.median(numeric_only=True)
```

# MODA

# MODA USANDO NUMPY

```python
values, counts = np.unique(penguins_df[x], return_counts=True)
mode_index = np.argmax(counts)
mode_value = values[mode_index]
print(mode_value)
```

## MODA USANDO PANDAS

```python
penguins_df[x].mode()[0]
```

## MODA DE TODAS LAS VARIABLES NUMERICAS

```python
penguins_df.mode(numeric_only=True)
```

# GRAFICAMOS MEDIDAS DE TENDENCIA CENTRAL DE VARIABLE BODY_MASS_G

```python
variable = 'body_mass_g'

sns.histplot(
    data=penguins_df,
    x=variable
)

# MEDIA
plt.axvline(
    x=penguins_df[variable].mean(),
    label='Media',
    color='red',
    linestyle='dashed',
    linewidth=2
)

# MEDIANA
plt.axvline(
    x=penguins_df[variable].median(),
    label='Mediana',
    color='green',
    linestyle='dashed',
    linewidth=2
)

#MODA
plt.axvline(
    x=penguins_df[variable].mode().values[0],
    label='Moda',
    color='yellow',
    linestyle='dashed',
    linewidth=2
)
plt.legend()

plt.show()
```
