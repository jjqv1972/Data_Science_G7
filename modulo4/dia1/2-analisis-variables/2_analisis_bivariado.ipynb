## 🔗 Análisis Bivariado

### 🔹 Definición
El **análisis bivariado** examina la relación entre dos variables, buscando asociaciones o dependencias.

### 🔹 Tipos de relaciones
1. **Numérica vs Numérica:** se evalúa correlación.
2. **Numérica vs Categórica:** se comparan medias o medianas.
3. **Categórica vs Categórica:** se usa una tabla de contingencia y pruebas de independencia (como Chi-cuadrado).

### 🔹 Correlación
La **correlación** mide la fuerza y dirección de una relación lineal entre dos variables numéricas.

- Rango: −1 ≤ *r* ≤ 1  
  - *r* ≈ 1 → relación positiva fuerte  
  - *r* ≈ 0 → sin relación lineal  
  - *r* ≈ −1 → relación negativa fuerte  

**Coeficiente de Pearson:**
> r = Σ((xᵢ−x̄)(yᵢ−ȳ)) / √(Σ(xᵢ−x̄)² × Σ(yᵢ−ȳ)²)

### 🔹 Visualizaciones típicas
- **Scatterplot (dispersión):** muestra la relación entre dos variables numéricas.  
- **Heatmap:** representa correlaciones múltiples en una matriz de colores.  
- **Barplot / Boxplot:** compara grupos según una variable categórica.

# VARIABLE CATEGORICA VS CATEGORICA

## Prueba Chi-cuadrado de Independencia

Determina si dos variables **categóricas** están asociadas.

**Hipótesis:**
- H₀: no hay relación (independencia).
- H₁: hay relación (dependencia).

**Criterio:**
- Si **p-value < 0.05** → existe relación significativa.
- Si **p-value ≥ 0.05** → las variables son independientes.

✅ **Valor óptimo:**  
p < 0.05 → asociación estadística entre las categorías.

```python
from scipy.stats import chi2_contingency
cat1 = 'species'
cat2 = 'sex'
```

## 1 - CREAMOS TABLA DE CONTINGENCIA

```python
import pandas as pd
tabla = pd.crosstab(penguins_df[cat1],penguins_df[cat2])
tabla
```

## 2 - CREAMOS GRÁFICO DE BARRAS APILADAS

```python
tabla.plot(kind='bar',stacked=True)
plt.title(f'Distribución conjunta de {cat1} y {cat2}')
plt.xlabel(cat1)
plt.ylabel("Frecuencia")
plt.legend(title=cat2)
plt.show()
```

## 3. Prueba de Chi-cuadrado de independencia

```python
chi2,p,dof,excepted = chi2_contingency(tabla)
print("\n🧪 Prueba Chi-cuadrado de independencia")
print(f"Chi-cuadrado: {chi2:.3f}")
print(f"Grados de libertad: {dof}")
print(f"p-value: {p:.5f}")
```

```python
if p < 0.05:
    interpretacion_chi = "Existe evidencia de asociación entre las dos variables categóricas."
else:
    interpretacion_chi = "No hay evidencia suficiente de asociación; las variables pueden considerarse independientes."

print(f"🧩 Interpretación: {interpretacion_chi}")
```
