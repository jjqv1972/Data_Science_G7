# Conversión de valores faltantes implícitos a explícitos

## Valores faltantes implícitos

Son datos que en realidad representan una ausencia de información, pero no están marcados como nulos de manera formal.

Por ejemplo:

- una cadena vacía `""`
- un espacio en blanco `" "`
- textos como `"NA"`, `"N/A"`, `"null"`, `"missing"`, `"sin dato"`
- números que se usan como marcador, por ejemplo `0`, `-1`, `999`

Estos valores parecen datos válidos, pero en el contexto del problema significan que falta información.

---

## Valores faltantes explícitos

Son los valores que ya están representados de forma estándar como nulos, por ejemplo:

- `NaN` en pandas
- `None` en Python

Cuando los faltantes están explícitos, es mucho más fácil:

- detectarlos
- contarlos
- analizarlos
- imputarlos o eliminarlos

---

## ¿Por qué convertir faltantes implícitos en explícitos?

Porque si no se hace esta conversión:

- pandas no los reconoce como nulos
- los conteos de valores faltantes serán incorrectos
- los análisis estadísticos pueden salir mal
- los modelos de machine learning pueden verse afectados

Por eso, antes de tratar los faltantes, primero debemos identificar si hay valores implícitos y convertirlos a `NaN`.

---

## Ejemplos comunes de faltantes implícitos

En una columna de edad:

- `0`
- `-1`
- `999`

En una columna de texto:

- `""`
- `"NA"`
- `"Sin dato"`
- `"Desconocido"`

En todos estos casos, debemos evaluar si realmente significan ausencia de información.

---

## Estrategia general

El proceso suele ser:

1. Explorar los valores únicos de cada columna
2. Detectar códigos sospechosos
3. Confirmar si representan ausencia de dato
4. Reemplazarlos por `NaN`
5. Recién después aplicar una técnica de tratamiento:
   - eliminar filas
   - imputar con media, mediana o moda
   - usar técnicas más avanzadas



```python
import pandas as pd
import numpy as np

data = {
    "nombre": ["Ana", "Luis", "Carlos", "María", "José"],
    "edad": [25, 30, 28, 35, 40],
    "ciudad": ["Lima", "NA", "", "Arequipa", "Sin dato"]
}

df = pd.DataFrame(data)

print("Dataset original:")
print(df)

print("\nValores faltantes detectados inicialmente:")
print(df.isna().sum())
```

# Convertimos valores faltantes implicitos a explicítos

```python
df["ciudad"] = df["ciudad"].replace(["", "NA", "Sin dato"], np.nan)

print("\nDataset después de convertir faltantes implícitos:")
print(df)

print("\nValores faltantes detectados ahora:")
print(df.isna().sum())
```

# convertir códigos numéricos a NaN

```python
data = {
    "producto": ["A", "B", "C", "D", "E"],
    "precio": [10.5, 0, 25.0, 999, 18.5]
}

df2 = pd.DataFrame(data)

print("Dataset original:")
print(df2)

print("\nValores únicos en precio:")
print(df2["precio"].unique())
```

## Reemplazo de códigos especiales

```python
df2["precio"] = df2["precio"].replace([0, 999], np.nan)

print("\nDataset después del reemplazo:")
print(df2)

print("\nValores faltantes por columna:")
print(df2.isna().sum())
```

# detectar faltantes implícitos explorando valores únicos

```python
data = {
    "genero": ["M", "F", "NA", "F", "", "Masculino", "Sin dato"],
    "edad": [20, 22, -1, 30, 25, 28, 999]
}

df3 = pd.DataFrame(data)

print("Valores únicos en genero:")
print(df3["genero"].unique())

print("\nValores únicos en edad:")
print(df3["edad"].unique())
```

# Convertimos los faltantes implícitos

```python
df3["genero"] = df3["genero"].replace(["NA", "", "Sin dato"], np.nan)
df3["edad"] = df3["edad"].replace([-1, 999], np.nan)

print("\nDataset limpio:")
print(df3)

print("\nValores faltantes por columna:")
print(df3.isna().sum())
```

# EJEMPLO DE CONVERSIÓN CON VARIAS COLUMNAS

```python
data = {
    "nombre": ["Ana", "Luis", "Pedro", "Lucía"],
    "correo": ["ana@gmail.com", "", "NA", "lucia@gmail.com"],
    "telefono": ["999111222", "Sin dato", "", "987654321"],
    "edad": [20, 0, 25, 30]
}

df4 = pd.DataFrame(data)

print("Dataset original:")
print(df4)
```

```python
valores_implicitos_texto = ["", "NA", "Sin dato"]
valores_implicitos_numericos = [0]

df4[["correo", "telefono"]] = df4[["correo", "telefono"]].replace(valores_implicitos_texto, np.nan)
df4["edad"] = df4["edad"].replace(valores_implicitos_numericos, np.nan)

print("\nDataset después de limpiar faltantes implícitos:")
print(df4)

print("\nValores faltantes por columna:")
print(df4.isna().sum())
```
