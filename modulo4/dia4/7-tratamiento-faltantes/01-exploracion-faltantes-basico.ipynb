# 📊 Tratamiento de Datos Faltantes (Missing Data)

## 🧠 ¿Qué son los datos faltantes?

Los **datos faltantes** (missing values) son valores que no están presentes en un dataset.  
Generalmente se representan como:

- `NaN` (Not a Number)
- `None`
- Celdas vacías

Ejemplo:

| nombre | edad | salario |
|--------|------|--------|
| Ana    | 25   | 2000   |
| Luis   | NaN  | 1800   |
| Marta  | 30   | NaN    |

---

## ⚠️ ¿Por qué ocurren los datos faltantes?

Los datos faltantes pueden aparecer por varias razones:

### 1. Error humano
- No se ingresó información
- Formularios incompletos

### 2. Problemas técnicos
- Fallos en sensores o sistemas
- Errores en la recolección de datos

### 3. Datos no aplicables
- Ejemplo: una persona sin hijos → campo "número de hijos" vacío

### 4. Pérdida de información
- Datos eliminados o corruptos durante procesos ETL

```python
import pandas as pd
```

# CREAMOS UN DATAFRAME DE PRUEBA

```python
df = pd.DataFrame.from_dict(
    data= {
        "a":list('asdfasdfas'),
        "b":range(0,10)
    }
)
df
```

#GENERAMOS VALORES NULOS

```python
df.iloc[2:5,0] = None
df.iloc[6:7,1] = None
df
```

# VERIFICAMOS VALORES NULOS

```python
df.isna()
```

# CONTEO DE NULOS

```python
df.isna().sum()
```

# CONTEO TOTAL DE NULOS

```python
df.isna().sum().sum()
```

# CONTEO POR CASOS

```python
df.isna().pipe(
    lambda df_1: (
        df_1.sum()
        .reset_index(name='n_missing')
        .rename(columns={'index':'variable'})
        .assign(
            n_cases=len(df_1),
            pct_missing=lambda df_2: df_2.n_missing / df_2.n_cases * 100,
        )
    )
)
```
