```python
import re

def to_snake_case(name):
    # Insert an underscore before any uppercase letter that is not at the start of the string
    name = re.sub(r'(?<!^)(?=[A-Z])', '', name)
    return name.lower()

data_df.columns = [to_snake_case(col) for col in data_df.columns]
```

```python
data_df
```

```python
data_df.isnull().sum()
```

# CONVERSION DE TIPOS DE DATOS

```python
data_df.columns
```

```python
columns_to_convert = ['area_terreno', 'area_construida', 'area_comun', 'porcentaje_propiedad', 'base_imponible']

for col in columns_to_convert:
    if data_df[col].dtype == 'object':
        # Remove commas for 'base_imponible' and strip spaces for all object columns
        if col == 'base_imponible':
            data_df[col] = data_df[col].astype(str).str.replace(',', '', regex=False).str.strip()
        else:
            data_df[col] = data_df[col].astype(str).str.strip()

        # Convert to numeric, coercing errors to NaN
        data_df[col] = pd.to_numeric(data_df[col], errors='coerce')

    # Ensure the column is float64, fill NaN with 0 if created by 'coerce'
    data_df[col] = data_df[col].fillna(0).astype('float64')

print(data_df.dtypes)
```
