# Muestreo con Titanic

## Problema
Queremos estimar la edad promedio de los pasajeros del Titanic y cuantificar la incertidumbre.

## Enfoque
- Tomar una muestra
- Calcular la media
- Construir un intervalo de confianza
- Comparar con la población disponible

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
titanic = sns.load_dataset("titanic").dropna(subset=["age"])
titanic
```

```python
# Muestra aleatoria
muestra = titanic.sample(n=120, random_state=42)

media = muestra["age"].mean()
s = muestra["age"].std()
n = len(muestra)
```

```python
from scipy.stats import norm
```

```python
z = norm.ppf(0.975)  # 95%
margen = z * (s / np.sqrt(n))
```

```python
li = media - margen
ls = media + margen

print("Media muestral:", round(media, 2))
print("IC 95%:", (round(li, 2), round(ls, 2)))
```

```python
print("Media edad población:", round(titanic["age"].mean(), 2))
```

```python
# Bootstrap para edad
bootstrap_means = []

for _ in range(2000):
    bs = muestra.sample(n=len(muestra), replace=True)
    bootstrap_means.append(bs["age"].mean())

bootstrap_means = pd.Series(bootstrap_means)

print("IC bootstrap 95% edad:",
      (round(bootstrap_means.quantile(0.025), 2),
       round(bootstrap_means.quantile(0.975), 2)))
```

```python
plt.figure(figsize=(8,5))
sns.histplot(bootstrap_means, kde=True)
plt.axvline(titanic["age"].mean(), color="red", linestyle="--", label="Media población")
plt.axvline(muestra["age"].mean(), color="black", label="Media muestra")
plt.title("Bootstrap de la edad media - Titanic")
plt.legend()
plt.show()
```
