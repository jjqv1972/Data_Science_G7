# Población, muestra y muestreo aleatorio

## ¿Qué es una población?
La **población** es el conjunto completo de datos que queremos estudiar.

## ¿Qué es una muestra?
La **muestra** es un subconjunto de la población. En análisis de datos, muchas veces no trabajamos con toda la población, sino con una muestra.

## ¿Qué es el muestreo?
El **muestreo** es el proceso de seleccionar observaciones desde una población.

## Muestreo aleatorio simple
En un **muestreo aleatorio simple**, cada elemento tiene la misma probabilidad de ser elegido.

## ¿Por qué importa?
Una buena muestra debe ser lo más representativa posible de la población. Si la muestra está sesgada, las conclusiones también lo estarán.

# CREAMOS UNA MUESTRA DE LA POBLACIÓN

```python
muestra_30 = penguins_df.sample(n=30,random_state=42)
print(f"Tamaño de la muestra {len(muestra_30)}")
```

El parámetro `random_state` en funciones que involucran aleatoriedad, como `sample()`, es crucial para la reproducibilidad de tus resultados. Te explico su significado:

Cuando realizas operaciones aleatorias (como seleccionar una muestra al azar), el resultado puede variar cada vez que ejecutas el código. Esto puede dificultar la depuración y la comparación de resultados. Aquí es donde entra `random_state`.

*   **¿Qué hace?** `random_state` establece una 'semilla' para el generador de números aleatorios. Si usas el mismo valor para `random_state` (por ejemplo, `42`), la secuencia de números aleatorios generada será la misma cada vez que ejecutes el código. Esto significa que si tú o cualquier otra persona ejecuta el código con `random_state=42`, obtendrá exactamente la misma muestra aleatoria.

*   **¿Por qué '42'?** El número `42` es una convención común en la comunidad de programación y ciencia de datos, a menudo se usa como un número arbitrario para `random_state` debido a la referencia de 'La Guía del Autoestopista Galáctico' donde 42 es la 'respuesta a la pregunta definitiva sobre la vida, el universo y todo'. Podrías usar cualquier otro número entero, como `0`, `1`, `100`, etc., y el efecto sería el mismo: reproducibilidad. Lo importante es que uses *el mismo* número si quieres resultados idénticos en ejecuciones distintas.

En resumen, `random_state=42` garantiza que la `muestra_30` que obtuviste sea siempre la misma cada vez que ejecutes ese código, haciendo tu trabajo reproducible.

# Comparar media poblacional vs media muestral

```python
media_poblacion = penguins_df["body_mass_g"].mean()
media_muestra = muestra_30["body_mass_g"].mean()

print("Media población:", round(media_poblacion, 2))
print("Media muestra:", round(media_muestra, 2))
```
