## ¿Qué es la correlación entre variables?

La **correlación** mide qué tan relacionada está una variable con otra. Es decir, si cuando una variable cambia, la otra también cambia de forma consistente.

### Por ejemplo:

- ✅ Si cuando estudias más, sacas mejores notas → **correlación positiva**
- ❌ Si cuando ves más televisión, sacas peores notas → **correlación negativa**
- ⚠️ Si no hay un patrón claro entre ver televisión y las notas → **correlación cercana a cero**

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0,100,100)
y = x ** 2
sns.scatterplot(x=x,y=y)
```

## ¿Qué es la correlación de Pearson?

La **correlación de Pearson** es una forma específica de medir la relación entre dos variables **numéricas**.

- Se representa con la letra **r** (coeficiente de correlación).
- Su valor está entre **–1 y 1**:

### Interpretación:

- **+1**: relación positiva perfecta (si una sube, la otra también sube proporcionalmente)
- **0**: sin relación lineal
- **–1**: relación negativa perfecta (si una sube, la otra baja proporcionalmente)

# MEDIR EL INDICE DE CORRELACIÓN DE PEARSON CON PYTHON

```python
df = pd.DataFrame(np.corrcoef(x,y),columns=['x','y'],index=['x','y'])
df
```

```python
sns.heatmap(
    data= df,
    cmap=sns.diverging_palette(20,230,as_cmap=True),
    center = 0,
    vmin = -1,
    vmax = 1,
    linewidths=0.5,
    cbar_kws={'shrink':0.5},
    annot=True
)
```
