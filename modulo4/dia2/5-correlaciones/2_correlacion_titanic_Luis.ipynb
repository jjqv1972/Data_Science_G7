# CORRELACIONES CON DATAFRAMES USANDO DATASET TITANIC


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
sns.set_style(style='whitegrid')
sns.set_context(context='notebook')
plt.rcParams['figure.figsize'] = (11, 9.4)


titanic_df = sns.load_dataset('titanic')
titanic_df.dropna(inplace=True)
titanic_df
```

## PARA VER LA CORRELACIÓN ENTRE 2 VARIABLES

```python
x = titanic_df['age']
y = titanic_df['fare']
np.corrcoef(x,y)
```

# MOSTRAMOS LA CORRELACION ENTRE TODAS LAS VARIABLES

```python
titanic_df.corr(numeric_only=True)
```

```python
sns.heatmap(
    data=titanic_df.corr(numeric_only=True),
    cmap=sns.diverging_palette(20, 220, as_cmap=True),
    center=0,
    vmin=-1,
    vmax=1,
    annot=True,
    linewidths=0.5,
    cbar_kws={'shrink': 0.5}
)
```
