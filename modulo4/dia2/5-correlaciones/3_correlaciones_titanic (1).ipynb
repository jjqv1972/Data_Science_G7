```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
sns.set_style(style='whitegrid')
sns.set_context(context='notebook')
plt.rcParams['figure.figsize'] = (11, 9.4)

titanic_df = sns.load_dataset('titanic')
titanic_df.dropna(inplace=True)
titanic_df
```

## PARA VER LA CORRELACIÓN ENTRE 2 VARIABLES FARE Y PCLASS

```python
x = titanic_df['fare']
y = titanic_df['pclass']
np.corrcoef(x,y)
```

# MOSTRAMOS LA CORRELACION ENTRE TODAS LAS VARIABLES

```python
titanic_df.corr(numeric_only=True)
```

```python
sns.heatmap(
    data=titanic_df.corr(numeric_only=True),
    cmap=sns.diverging_palette(20, 220, as_cmap=True),
    center=0,
    vmin=-1,
    vmax=1,
    annot=True,
    linewidths=0.5,
    cbar_kws={'shrink': 0.5}
)
```

## MAPA DE CALOR PARA UNA SOLA VARIABLE DEPENDIENTE

```python
corr_matrix = titanic_df.corr(numeric_only=True)
corr_body_mass = corr_matrix[['survived']].sort_values(by='survived',ascending=True)
sns.heatmap(
    data=corr_body_mass,
    cmap=sns.diverging_palette(20, 230, as_cmap=True),
    center=0,
    vmin=-1,
    vmax=1,
    linewidths=0.5,
    cbar_kws={'shrink': 0.5},
    annot=True
)
```
