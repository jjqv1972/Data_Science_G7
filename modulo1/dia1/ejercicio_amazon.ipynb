```python
import pandas as pd

df = pd.read_csv('amazon.csv')
df.head(10)
     
```

```python
# Limpiar y convertir 'rating_count' a numérico
df['rating_count'] = df['rating_count'].str.replace(',', '', regex=True)
df['rating_count'] = pd.to_numeric(df['rating_count'], errors='coerce')

# Eliminar filas donde 'rating_count' sea NaN después de la conversión
df.dropna(subset=['rating_count'], inplace=True)

# Agrupar por producto y sumar el rating_count para identificar los 'más vendidos'
top_products = df.groupby('product_name')['rating_count'].sum().sort_values(ascending=False).head(10)

import matplotlib.pyplot as plt
import seaborn as sns

# Crear el gráfico de barras
plt.figure(figsize=(12, 8))
sns.barplot(x=top_products.values, y=top_products.index, palette='viridis')
plt.title('Top 10 Productos más vendidos (por número de valoraciones)')
plt.xlabel('Número total de valoraciones')
plt.ylabel('Nombre del producto')
plt.show()
```
