## 🔳 ¿Qué es un subplot?

Un **subplot** es una técnica que permite dibujar **múltiples gráficos** dentro de una misma figura en `Matplotlib`.

Esto es útil cuando queremos comparar diferentes visualizaciones lado a lado o apilar varias gráficas en una misma vista.

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0,5,11)
y = x**2
```

# CREAMOS 2 GRAFICOS (SUBPLOTS)

```python
plt.subplot(1,2,1)
plt.plot(x,y,'b--')

plt.subplot(1,2,2)
plt.hist(y)

plt.show()
```

# SUBPLOT CON VARIOS GRAFICOS

```python
plt.subplot(1,2,1)
plt.plot(x,y,'b--')
plt.plot(y,x,'r:')

plt.subplot(1,2,2)
plt.pie(x)

plt.show()
```

# CAMBIAMOS ORIENTACIÓN DE LOS GRAFICOS

```python
plt.subplot(2,1,1)
plt.plot(x,y,'b--')
plt.plot(y,x,'r:')

plt.subplot(2,1,2)
plt.hist(x)

plt.show()
```
