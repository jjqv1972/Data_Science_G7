# EJERCICIOS CON MATPLOTLIB

```python
import pandas as pd
df_books = pd.read_csv('https://github.com/cesarmayta/datasets/raw/refs/heads/main/libros.csv')
df_books.head(2)
```

### x = Year
### y = Precio Promedio

```python
df_eje1 = df_books.groupby('Year').agg({'Price':'mean'}).reset_index()
df_eje1
```

```python
x = df_eje1['Year']
y = df_eje1['Price']
```

📌 **Ejercicio 1: Evolución del precio promedio por año**

Muestra cómo ha cambiado el precio promedio de los libros en el tiempo.

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(x,y,color='green',label='Precio promedio de libros por año')
plt.title('Evolución del precio promedio de libros por año')
plt.xlabel('Año')
plt.ylabel('Precio promedio')
plt.legend()
plt.show()
```

📌 **Ejercicio 2: Comparar calificación(Rating) promedio por género**
Compara la calificación promedio entre libros de ficción y no ficción, mostrar un grafico de barras

```python
df_eje2 = df_books.groupby('Genre')['User Rating'].mean().reset_index()
df_eje2
```

```python
x = df_eje2['Genre']
y = df_eje2['User Rating']

plt.bar(x,y,color=['blue','red'],label='calificación promedio por género')
plt.title('Comparación de calificación promedio por género')
plt.xlabel('Género')
plt.ylabel('Calificación promedio')

plt.legend()
plt.show()
```
