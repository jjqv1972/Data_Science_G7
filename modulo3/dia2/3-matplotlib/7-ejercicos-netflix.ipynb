## 🧪 Ejercicio 1: Histograma de años

**Objetivo:**  
Analizar cómo se distribuyen los años de lanzamiento del contenido en Netflix.

**Instrucciones:**
- Cargar el dataset `netflix_titles.csv`
- Crear un histograma usando la columna `release_year`
- Agregar título al gráfico
- Agregar etiquetas a los ejes

**Preguntas:**
- ¿En qué años hay más contenido?
- ¿La mayoría del contenido es reciente o antiguo?




```python
x = df['release_year']

plt.hist(x,color='green')
plt.title('Distribución de los años de lanzamiento de contenido en Netflix')
plt.xlabel('Año de lanzamiento')
plt.ylabel('Frecuencia')
plt.legend()
plt.show()
```

## 🧪 Ejercicio 2: Gráfico Scatter (Año vs Duración)

**Objetivo:**  
Analizar la relación entre el año de lanzamiento y la duración de las películas.

**Instrucciones:**
- Filtrar solo los registros donde `type` sea "Movie"
- Convertir la columna `duration` a valores numéricos (eliminar texto "min")
- Crear un gráfico scatter:
  - Eje X: `release_year`
  - Eje Y: `duration`
- Agregar título y etiquetas

**Preguntas:**
- ¿Las películas más recientes duran más o menos?
- ¿Existen valores extremos (películas muy largas)?


```python
df.info()
```

```python
df.isnull().sum()
```

```python
df.dropna(inplace=True)
```

```python
movies = df[df['type'] == 'Movie']
movies
```

```python
movies['duration'] = movies['duration'].str.replace(' min', '').astype(int)
movies['duration']
```

```python
x = movies['release_year']
y = movies['duration']

plt.scatter(x,y,color='red')
plt.title('Relación entre el año de lanzamiento y la duración de las películas')
plt.xlabel('Año de lanzamiento')
plt.ylabel('duración')
plt.legend()
plt.show()
```
