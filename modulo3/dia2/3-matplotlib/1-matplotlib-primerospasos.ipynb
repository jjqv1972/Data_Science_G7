## 🔷 ¿Qué es Matplotlib?
**Matplotlib** es una biblioteca de visualización en Python que permite crear gráficos 2D de alta calidad. Es ampliamente usada en ciencia de datos, machine learning y análisis exploratorio.

Se inspira en MATLAB y ofrece una interfaz orientada a objetos para un control más preciso de los gráficos.

# IMPORTAMOS EL PAQUETE PYPLOT

````python
import matplotlib.pyplot as plt
````

# GENERAMOS DATOS PARA NUESTRO PRIMER PLOT

````python
import numpy as np
x = np.linspace(0,5,11)
x
````

````python
y = x**2
y
````

# CREAMOS NUESTRO PRIMER PLOT

````python
plt.plot(x,y)
plt.show()
````

# CAMBIAMOS EL COLOR DEL GRAFICO

````python
plt.plot(x,y,'r')
plt.show()
````

````python
plt.plot(x,y,'go')
plt.show()
````

````python
plt.plot(x,y,'bo-')
plt.show()
````

````python
plt.plot(x,y,'gD:')
plt.show()
````

````python
plt.plot(x,y,color='#b942f5',linestyle='--',marker='o')
plt.show()
````

### 🎨 Parámetros

- **`color`**: `'red'`, `'blue'`, `'#FF5733'`
- **`linestyle`**: `'-'`, `'--'`, `':'`, `'-.`
- **`marker`**: `'o'`, `'s'`, `'*'`, `'x'`,
