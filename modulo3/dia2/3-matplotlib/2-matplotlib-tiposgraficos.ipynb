# TIPOS DE GRAFICOS CON MATPLOTLIB

```python
import matplotlib.pyplot as plt
```

## 📈 1. Gráfico de líneas

Se utiliza para visualizar tendencias o evolución de datos en el tiempo.

```python
x = [1, 2, 3, 4, 5]
y = [10, 20, 25, 30, 35]
plt.plot(x,y,marker='o',linestyle='-',color='blue')
plt.show()
```

## 📊 2. Gráfico de barras

Útil para comparar cantidades entre distintas categorías.

```python
categorias = ['Enero', 'Febrero', 'Marzo', 'Abril']
valores = [40, 70, 30, 85]
```

```python
plt.bar(categorias,valores,color='orange')
plt.show()
```

```python
plt.barh(categorias,valores,color='green')
plt.show()
```

## 🧁 4. Gráfico de pastel (pie chart)

Muestra proporciones relativas de un total.

```python
etiquetas = ['Manzanas', 'Plátanos', 'Peras']
tamaños = [30, 50, 20]
```

```python
plt.pie(tamaños,labels=etiquetas,startangle=90,autopct='%1.1f%%')
plt.show()
```

## 🔘 5. Gráfico de dispersión (scatter plot)

Ideal para observar correlaciones entre dos variables numéricas.

```python
import numpy as np

x = np.random.rand(50)
y = np.random.rand(50)
```

```python
plt.scatter(x,y,color='red')
plt.show()
```

## 📊 6. Histograma

Muestra la distribución de frecuencia de una variable continua.

```python
datos = np.random.randn(1000)
datos
```

```python
plt.hist(datos,bins=30,color='orange',edgecolor='black')
plt.show()
```

## 📦 7. Diagrama de caja (boxplot)

Útil para detectar valores atípicos y ver la distribución.

```python
data = [np.random.normal(0,std,100) for std in range(1,4)]
data
```

```python
plt.boxplot(data,vert=True,patch_artist=True)
plt.show()
```
