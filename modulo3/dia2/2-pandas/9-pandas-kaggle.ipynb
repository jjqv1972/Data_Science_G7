### Ejercicio 1: Filtrado de datos 🔍

1. Filtra todas las películas que tengan:
   - Un rating mayor o igual a 9.0
   - Y que sean del año 2000 en adelante

2. Del resultado, muestra solamente estas columnas:
   - `Rank`, `Title`, `Year`, `Rating`, `Certificate`, `Genres`

3. Ordena el resultado por `Rating` (descendente) y luego por `Year` (descendente).

```python
df_movies[
    (df_movies['Rating'] >= 9.0) &
    (df_movies['Year'] >= 2000)
]
```

```python
cols = ['Rank','Title','Year','Rating','Certificate','Genres']
```

```python
df_movies[
    (df_movies['Rating'] >= 9.0) &
    (df_movies['Year'] >= 2000)
][cols]
```

```python
df_movies[
    (df_movies['Rating'] >= 9.0) &
    (df_movies['Year'] >= 2000)
][cols].sort_values(by=['Rating','Year'],ascending=[False,False])
```

### Ejercicio 2: Agrupaciones 📊

1. Agrupa las películas por año (`Year`) y calcula:
   - El número de películas por año.
   - El rating promedio por año.

2. Ordena el resultado de mayor a menor rating promedio.

3. Muestra solo las primeras 10 filas del resultado final.

```python
df_movies.groupby('Year').agg(
    peliculas_año=('Title', 'count'),
    promedio_anual=('Rating', 'mean')
).sort_values(by='promedio_anual', ascending=False).head(10).reset_index()
```
