# 🐼 Introducción a Pandas en Python

Pandas es una de las librerías más populares de Python para **análisis y manipulación de datos**.  
Está construida sobre NumPy y proporciona estructuras de datos flexibles y eficientes, principalmente: Pandas Series y Pandas DataFrame

# Pandas Series

```python
import pandas as pd
```

```python
paises = pd.Series(['Argentina', 'Peru', 'Chile', 'Paraguay', 'Uruguay'])
paises
```

```python
type(paises)
```

```python
paises = pd.Series(['Perú','Chile','Colombia','Ecuador'],index=[1,3,5,7])
paises
```

## CREAMOS PANDAS SERIES DESDE UN DICCIONARIO

```python
dict = {
    1:'Perú',
    2:'Chile',
    3:'Colombia',
    4:'Ecuador'
}
pd.Series(dict)
```

# Pandas DATAFRAME

```python
dict = {
    'pais':['Perú','Chile','Colombia','Ecuador'],
    'pbi':[183,170,160,150],
    'habitantes':[33,22,22,50]
}
df_paises = pd.DataFrame(dict,index=[1,7,10,30])
df_paises
```
