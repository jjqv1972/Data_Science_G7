# 🧼 Tratamiento de Nulos y Duplicados en PySpark

```python
import seaborn as sns
from pyspark.sql import SparkSession

# Inicializar SparkSession
spark = SparkSession.builder.appName("TitanicAnalysis").getOrCreate()

# Cargar el dataset de Titanic desde seaborn (primero lo cargamos a pandas)
df = sns.load_dataset('titanic')

# Crear un DataFrame de PySpark a partir del DataFrame de Pandas
df = spark.createDataFrame(df)

# Mostrar el esquema y las primeras filas del DataFrame de PySpark
df.printSchema()
df.show()
```

```python
df.count()
```

# dropna

```python
df.count() - df.dropna(how='any').count()
```

```python
# Eliminar filas si todas las columnas están nulas
df.count() - df.dropna(how='all').count()
```

```python
df.dropna(subset=['survived', 'pclass']).count()
```

# dropduplicates

```python
df.dropDuplicates().count()
```

```python
df.dropDuplicates(['survived', 'pclass']).count()
```

# 💾 Persistencia de DataFrames en PySpark

## 🔹 ¿Qué significa "persistencia" en PySpark?

En PySpark, los DataFrames son **inmutables**. Esto significa que **cualquier transformación** que se realiza sobre un DataFrame (como `dropna()`, `filter()`, `withColumn()`, etc.) **no modifica el DataFrame original**, sino que devuelve **un nuevo DataFrame modificado**.

```python
df.count()
```

```python
df_drop = df.dropDuplicates()
df_drop.count()
```
