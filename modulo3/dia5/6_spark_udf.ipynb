## 📘 Capítulo 1: ¿Qué son las UDF?

### 🔹 Definición

Una **UDF (User Defined Function)** es una función que tú defines en Python (o Scala/Java) y que puedes usar dentro de PySpark para aplicar lógica personalizada a las columnas de un DataFrame.

```python
from pyspark.sql import SparkSession
from pyspark.sql.functions import udf
from pyspark.sql.types import StringType
```

```python
# Crear SparkSession
spark = SparkSession.builder.appName("UDFLibros").getOrCreate()

# Leer archivo libros.csv (sube el archivo en la barra lateral de Colab)
df = spark.read.option("header", "true").option("inferSchema", "true").csv("libros.csv")
df.show(5)
```

# Definir una función en Python

```python
def clasificar_rating(rating):
  if rating >= 4.5:
    return "Excelente"
  elif rating >= 4.0:
    return "Bueno"
  else:
    return "Regular"
```

# Registrar la UDF en PySpark

```python
clasificar_rating_udf = udf(clasificar_rating, StringType())
```

# Aplicar la UDF sobre el DataFrame

```python
df_resultado = df.withColumn("Clasificación",clasificar_rating_udf(df["User Rating"]))
df_resultado.select("Name","User Rating","Clasificación").show(10)
```
