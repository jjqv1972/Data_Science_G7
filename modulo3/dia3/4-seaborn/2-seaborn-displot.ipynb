### 📊 `displot` en Seaborn

`displot` (abreviatura de *distribution plot*) es una función de **Seaborn** que se usa para **visualizar distribuciones de datos**.  
Es una función de **alto nivel** que combina varias capacidades como:

- Histogramas
- Gráficos de densidad (`kde`)
- Facetas (múltiples gráficos en una sola figura)

Gracias a su flexibilidad, `displot` permite crear visualizaciones complejas de forma sencilla y organizada.

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

# CARGAMOS DATASET DE MUESTRA TIPS

```python
tips = sns.load_dataset('tips')
tips
```

# CREAMOS GRAFICO CON DISPLOT

```python
sns.displot(data=tips,x='tip')
plt.show()
```

# RELACIÓN ENTRE 2 VARIABLES CON DISPLOT

```python
sns.displot(data=tips,x='total_bill',y='tip')
plt.show()
```

# AGRUPACIÓN DE DATOS

```python
sns.displot(data=tips,x='tip',hue='sex')
plt.show()
```

# GRAFICO DE DENSIDAD CON KDE

```python
sns.displot(data=tips,x='tip',hue='sex',kind='kde')
plt.show()
```

# AGREGAMOS LEYENDAS PROPIAS AL GRAFICO

```python
sns.displot(data=tips,x='tip',hue='sex',kind='kde')
plt.title('Propinas por sexo')
plt.xlabel('propina')
plt.ylabel('densidad')
plt.legend()
plt.show()
```
