## 🧩 ¿Qué son los datos categóricos?

Los **datos categóricos** son variables que representan **grupos, categorías o etiquetas** en lugar de cantidades numéricas.  
Estas variables indican **tipos o clases** y **no tienen un orden matemático** (aunque a veces pueden tener un orden lógico).

---

### 📌 Ejemplos comunes de datos categóricos:

- `sex`: "Male" o "Female"
- `smoker`: "Yes" o "No"
- `day`: "Thur", "Fri", "Sat", "Sun"
- `time`: "Lunch" o "Dinner"


```python
import seaborn as sns
import matplotlib.pyplot as plt

tips = sns.load_dataset('tips')
tips
```

# CREAMOS GRAFICOS CON DATOS CATEGORICOS

## 🔢 ¿Qué es `countplot` en Seaborn?

`countplot` es una función de **Seaborn** diseñada para visualizar **la cantidad (conteo) de observaciones** en cada categoría de una variable categórica.

Es útil cuando quieres saber **cuántos elementos hay en cada grupo o clase** dentro de una columna categórica.

---

```python
sns.countplot(data=tips,x='day',hue='sex')
plt.show()
```

# STRIPPLOT

## 🎯 ¿Qué es `stripplot` en Seaborn?

`stripplot` es una función de **Seaborn** que se utiliza para **visualizar la distribución de datos categóricos frente a una variable numérica**, colocando **cada observación individual como un punto** en el gráfico.

Es ideal para ver cómo se **dispersan los datos** en cada categoría.

```python
sns.stripplot(data=tips,x='day',y='tip',hue='sex',dodge=True)
plt.show()
```

# CATPLOT

`catplot` (abreviatura de *categorical plot*) es una **función de alto nivel** en Seaborn que permite crear **gráficos categóricos flexibles y combinables** fácilmente.

Es una forma **generalizada** de graficar relaciones entre **variables categóricas** y **variables numéricas**, y puede generar varios tipos de gráficos dependiendo del parámetro `kind`.

```python
sns.catplot(data=tips,x='day',col='time',hue='sex',kind='count')
plt.show()
```

```python
sns.catplot(data=tips,x='day',y='total_bill',kind='box')
plt.suptitle("Total de la cuenta por día (boxplot)", y=1.02)
plt.show()
```

### 🔧 Parámetros principales de `catplot`

- `x`: Variable **categórica**
- `y`: Variable **numérica**
- `kind`: Tipo de gráfico. Valores posibles:
  - `"strip"`: puntos individuales (por defecto)
  - `"swarm"`: puntos no superpuestos
  - `"box"`: diagrama de caja
  - `"violin"`: distribución + box
  - `"bar"`: barra (muestra el promedio)
  - `"point"`: puntos conectados (con líneas)
  - `"count"`: conteo de categorías (no necesita `y`)
- `hue`, `col`, `row`: Para **crear subgráficos por grupo** y comparar visualmente diferentes categorías
