# 📊 Boxplot (Diagrama de Caja y Bigotes)

## 🧠 ¿Qué es un Boxplot?

Un **boxplot** (diagrama de caja y bigotes) es una herramienta de visualización que permite analizar la **distribución de un conjunto de datos numéricos**.

Es muy utilizado en análisis de datos porque resume información clave de forma visual.

---

## 🎯 ¿Para qué sirve?

Un boxplot permite:

- Visualizar la **mediana** de los datos
- Identificar la **dispersión**
- Detectar **valores atípicos (outliers)**
- Comparar distribuciones entre diferentes grupos

---

## 📦 Componentes de un Boxplot

Un boxplot está compuesto por:

- **Caja (Box):**
  Representa el rango intercuartílico (IQR), es decir, el 50% central de los datos (entre Q1 y Q3)

- **Mediana (Q2):**
  Línea dentro de la caja que indica el valor central de los datos

- **Bigotes (Whiskers):**
  Extensión que muestra el rango de los datos sin considerar outliers

- **Outliers:**
  Puntos que están fuera del rango normal de los datos

---

## 📐 Conceptos Estadísticos Clave

- **Q1 (Primer cuartil):** 25% de los datos están por debajo
- **Q2 (Mediana):** 50% de los datos
- **Q3 (Tercer cuartil):** 75% de los datos
- **IQR (Rango intercuartílico):**
  
  IQR = Q3 - Q1



```python
import seaborn as sns
import matplotlib.pyplot as plt

tips = sns.load_dataset('tips')
tips.head()
```

# Boxplot básico

```python
sns.boxplot(data=tips, y='tip')
plt.show()
```

# Boxplot por categoría

```python
sns.boxplot(data=tips,x='day',y='tip')
plt.show()
```

# Usando hue (comparación dentro de grupos)

```python
sns.boxplot(data=tips,x='day',y='tip',hue='sex')
plt.show()
```

# Personalizando el gráfico

```python
sns.boxplot(
    x='day',
    y='tip',
    hue='sex',
    data=tips,
    palette='Set2'
)

plt.title('Propinas por Día y Género')
plt.xlabel('Día')
plt.ylabel('Propina')
plt.legend(title='Género')

plt.show()
```
