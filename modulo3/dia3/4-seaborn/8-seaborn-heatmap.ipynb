# 🔥 Heatmap (Mapa de Calor)

## 🧠 ¿Qué es un Heatmap?

Un **heatmap (mapa de calor)** es una visualización que representa valores numéricos mediante **colores**.

Los valores más altos y más bajos se muestran con diferentes intensidades o tonos, lo que permite identificar patrones rápidamente.

---

## 🎯 ¿Para qué sirve?

Un heatmap permite:

- Detectar **patrones y tendencias**
- Visualizar **relaciones entre variables**
- Identificar **zonas de alta o baja intensidad**
- Analizar **correlaciones** entre variables numéricas

---

## 🌈 ¿Cómo funciona?

Cada celda del heatmap:

- Representa un valor numérico
- Se colorea según su magnitud

Ejemplo:

- Colores claros → valores bajos
- Colores oscuros → valores altos


```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

tips = sns.load_dataset('tips')
tips.head()
```

# Heatmap de correlación

```python
corr = tips.corr(numeric_only=True)
corr
```

```python
sns.heatmap(corr,annot=True,cmap='coolwarm')
plt.show()
```

# Queremos ver el promedio de propinas por día y momento

```python
pivot = tips.pivot_table(
    values='tip',
    index='day',
    columns='time',
    aggfunc='mean'
)

pivot
```

```python
sns.heatmap(pivot, annot=True, cmap='YlGnBu')
plt.title('Propina Promedio por Día y Momento')
plt.show()
```

# Heatmap con conteo (frecuencia)

```python
pivot_count = tips.pivot_table(
    values='tip',
    index='day',
    columns='time',
    aggfunc='count'
)

sns.heatmap(pivot_count, annot=True, cmap='Blues')
plt.title('Cantidad de Registros por Día y Momento')
plt.show()
```
