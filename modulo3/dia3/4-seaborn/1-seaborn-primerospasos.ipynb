# SEABORN

**Seaborn** es una biblioteca de Python para visualización de datos, que:

- Está construida sobre **Matplotlib**.
- Integra muy bien con **Pandas DataFrames**.
- Facilita la creación de gráficos estadísticos complejos.
- Aplica automáticamente un estilo visual atractivo y moderno a los gráficos.

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

# EJEMPLO DE GRAFICO CON SEABORN

```python
x_array =['A','B','C']
y_array =[1,2,3]
```

```python
sns.set(style='darkgrid',palette='dark',font_scale=1)
```

```python
sns.barplot(x=x_array,y=y_array)
plt.show()
```

# GARGAMOS UN DATASET EN UN DATAFRAME DE PANDAS

```python
titanic_df = sns.load_dataset('titanic')
titanic_df.head(5)
```

```python
sns.barplot(data=titanic_df,x='sex',y='survived')
plt.show()
```
