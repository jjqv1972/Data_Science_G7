```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
x = list(range(1, 8))
y = [i * 2 for i in x]
```

```python
x
```

```python
y
```

```python
plt.scatter(x,y)
plt.show()
```

# REGRESIÓN LINEAL CON SKLEARN

```python
from sklearn.linear_model import LinearRegression
```

```python
x = np.array(x).reshape(-1,1)
```

## ENTRENAMOS EL MODELO

```python
model = LinearRegression()
model.fit(x,y)
```

## PROBAMOS CON NUEVOS VALORES

```python
new_x = np.array([8,9,10]).reshape(-1,1)

new_y_pred = model.predict(new_x)

print("nuevos valores de x : ",new_x.flatten())
print("predicciones de y : ",new_y_pred)
```

## GRAFICAMOS EL RESULTADO

```python
plt.scatter(x, y, color='green', label='Original Data')
plt.scatter(new_x.flatten(), new_y_pred, color='red', label='Predictions')
plt.xlabel("x")
plt.ylabel("y")
plt.title("Original Data and Predictions")
plt.legend()
plt.show()
```
