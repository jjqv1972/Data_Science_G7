# 📘 Regresión LASSO (Least Absolute Shrinkage and Selection Operator)

La **Regresión LASSO** es una variante de la regresión lineal que utiliza una **regularización L1** para evitar el sobreajuste y mejorar la interpretabilidad del modelo.  
Su objetivo principal es **reducir la magnitud de los coeficientes**, e incluso **llevar algunos exactamente a cero**, logrando una forma de **selección automática de variables**.

---

## 🎯 Objetivo de LASSO

LASSO busca resolver el siguiente problema de optimización:

$$
\hat{\beta} = \arg\min_{\beta}\left\{
\sum_{i=1}^{n}(y_i - \hat{y}_i)^2 +
\lambda \sum_{j=1}^{p}|\beta_j|
\right\}
$$

donde:

- $y_i$ = valor real  
- $\hat{y}_i$ = predicción del modelo  
- $\beta_j$ = coeficientes del modelo  
- $\lambda$ = parámetro de regularización  
- $|\beta_j|$ = penalización L1

---

## 🔍 ¿Qué hace la penalización L1?

La penalización L1 tiene dos efectos:

### **1️⃣ Reduce la magnitud de los coeficientes (shrinkage)**
Evita que el modelo tenga valores extremos que pueden ser causados por ruido.

### **2️⃣ Lleva algunos coeficientes exactamente a cero**
Esto convierte a LASSO en un método de:
- **Regularización**
- **Selección de características**

El modelo final queda solo con las variables más relevantes.

---

## 🧠 ¿Cuándo utilizar LASSO?

| Situación | ¿LASSO ayuda? |
|----------|----------------|
| Muchas variables irrelevantes | ✅ Excelente |
| Se quiere seleccionar características | ✅ Sí |
| Existen variables muy correlacionadas | ⚠️ Puede elegir una y descartar otras |
| Datos con multicolinealidad | 👍 Funciona, pero Ridge es mejor |
| Dataset pequeño | 👍 Recomendable |

---

```python
import numpy as np
from sklearn.linear_model import Lasso
```

```python
x = np.array([1, 2, 3, 4])
y = np.array([2, 4, 5, 4])
```

```python
X = x.reshape(-1, 1)
```

```python
lasso = Lasso(alpha=0.5) # alpha = λ de la penalización
```

```python
lasso.fit(X,y)
```

```python
# Coeficientes obtenidos
print("Beta0 (intercept):", lasso.intercept_)
print("Beta1 (pendiente):", lasso.coef_[0])
```

```python
y_pred = lasso.predict(X)
print("Predicciones : ",y_pred)
```
