# APLICANDO LASSO A DATASET HOUSING

# 1 - IMPORTAMOS LIBRERIAS

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

# 2 - CARGAMOS DATASET

```python
df = pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/housing/housing.data',header=None,sep=r'\s+')
df.columns =  ['CRIM', 'ZN', 'INDUS', 'CHAS', 'NOX', 'RM', 'AGE', 'DIS', 'RAD', 'TAX', 'PTRATIO', 'B', 'LSTAT', 'MEDV']
df.head(5)
```

```python
corr_matrix = df.corr()
corr_medv = corr_matrix[['MEDV']].sort_values(by='MEDV',ascending=False)
corr_medv
```

```python
cols = corr_medv.index.tolist()
cols.remove('MEDV')
cols
```

# 3 - ENTRENAMIENTO CON LASSO

## 3.1 DEFINIMOS VARIABLE X y Y

```python
X = df[cols].values
y = df['MEDV'].values.reshape(-1,1)
```

## 3.2 DIVIDIMOS DATASET EN TRAIN Y TEST

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.30)
```

## 3.3 ESTANDARIZAMOS TODAS LAS VARIABLES

```python
from sklearn.preprocessing import StandardScaler
sc_x = StandardScaler()
X_train = sc_x.fit_transform(X_train)
X_test = sc_x.transform(X_test)

sc_y = StandardScaler()
y_train = sc_y.fit_transform(y_train)
y_test = sc_y.transform(y_test)
```

## 3.4 ENTRENAMOS EL MODELO

```python
from sklearn.linear_model import Lasso
from sklearn.metrics import mean_squared_error, r2_score

lasso = Lasso(alpha=0.1)
lasso.fit(X_train,y_train)

y_pred_lasso = lasso.predict(X_test)
```

```python
#evaluamos el modelo
mse_lasso = mean_squared_error(y_test,y_pred_lasso)
r2_lasso = r2_score(y_test,y_pred_lasso)

print(f'Lasso - Mean Squared Error (MSE): {mse_lasso:.4f}')
print(f'Lasso - R-squared (R2): {r2_lasso:.4f}')
```
