# 📘 Regresión Ridge (Regularización L2)

La **Regresión Ridge** es una extensión del modelo de regresión lineal que incorpora un término de **regularización L2** con el objetivo de reducir el sobreajuste, controlar la magnitud de los coeficientes y mejorar la estabilidad del modelo, especialmente cuando existe multicolinealidad entre las variables predictoras.

---

## 🎯 Objetivo de Ridge Regression

Ridge ajusta los coeficientes resolviendo el siguiente problema de optimización:

$$
\hat{\beta} = \arg\min_{\beta}\left\{
\sum_{i=1}^{n}(y_i - \hat{y}_i)^2 +
\lambda \sum_{j=1}^{p}\beta_j^2
\right\}
$$

donde:

- $y_i$ = valor real  
- $\hat{y}_i$ = predicción del modelo  
- $\beta_j$ = coeficientes del modelo  
- $\lambda$ = parámetro de regularización  
- $\beta_j^2$ = penalización L2  

---

## 🔍 ¿Qué hace la penalización L2?

La regularización L2 agrega un costo al modelo cuando los coeficientes tienen valores grandes.

### ✔ Efectos principales:

### **1️⃣ Reduce la magnitud de los coeficientes**
Evita que los parámetros crezcan de forma exagerada debido a ruido o correlaciones fuertes.

### **2️⃣ No lleva coeficientes a cero**
A diferencia de Lasso (L1), Ridge **nunca elimina variables**, solo disminuye su impacto.

### **3️⃣ Estabiliza el modelo ante multicolinealidad**
Cuando dos o más variables están altamente correlacionadas, la regresión lineal tradicional da coeficientes inestables.  
Ridge los suaviza y distribuye el peso entre ellas.

```python
import numpy as np
from sklearn.linear_model import Ridge
```

```python
x = np.array([1, 2, 3, 4])
y = np.array([2, 4, 5, 4])
```

```python
X = x.reshape(-1, 1)
```

```python
ridge = Ridge(alpha=0.5)
```

```python
ridge.fit(X,y)
```

# Coeficientes obtenidos

```python
print("Beta0 (intercept):", ridge.intercept_)
print("Beta1 (pendiente):", ridge.coef_[0])
```

# Predicciones

```python
y_pred = ridge.predict(X)
print("Predicciones:", y_pred)
```
