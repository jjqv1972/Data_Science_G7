# 🌳 Random Forest Regressor

## 1️⃣ ¿Qué es Random Forest para regresión?

**Random Forest Regressor** es un algoritmo de **aprendizaje supervisado** basado en **ensambles de árboles de decisión**.

📌 Idea principal:
- Entrena **muchos árboles**
- Cada árbol ve:
  - Un **subconjunto aleatorio de datos**
  - Un **subconjunto aleatorio de variables**
- La predicción final es el **promedio** de todos los árboles

$$
\hat{y}(x) = \frac{1}{T} \sum_{t=1}^{T} f_t(x)
$$

# 🌲 Hiperparámetros de Random Forest Regressor

| Parámetro           | Controla                              | Regla práctica                          |
| ------------------- | ------------------------------------- | --------------------------------------- |
| `n_estimators`      | Número de árboles                     | Empieza con 100                          |
| `max_depth`         | Profundidad máxima de cada árbol      | `None` o valores entre 10–30             |
| `min_samples_split` | Mínimo de muestras para dividir nodo  | 2 (default)                              |
| `min_samples_leaf`  | Mínimo de muestras en hoja            | 1–5 para reducir overfitting             |
| `max_features`      | Nº de variables por split             | `"sqrt"` (default recomendado)           |
| `bootstrap`         | Muestreo con reemplazo                | `True` (recomendado)                     |
| `criterion`         | Métrica de error                      | `"squared_error"`                        |
| `n_jobs`            | Paralelismo                           | `-1` (usar todos los núcleos)            |
| `random_state`      | Reproducibilidad                      | Fijar valor (ej. 42)                     |


```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.ensemble import RandomForestRegressor
```

```python
X = np.arange(1, 21).reshape(-1, 1)
y = 3 * X.ravel() + np.random.randn(20)
```

```python
plt.scatter(X,y)
plt.show()
```

```python
rf = RandomForestRegressor(
    n_estimators=200,
    max_depth=10,
    min_samples_leaf=2,
    random_state=42,
    n_jobs=-1
)

rf.fit(X, y)
```

```python
X_future = np.array([[21], [22], [23]])
y_future = rf.predict(X_future)

for x, y_pred in zip(X_future.ravel(), y_future):
    print(f"x = {x} → y ≈ {y_pred:.2f}")
```
